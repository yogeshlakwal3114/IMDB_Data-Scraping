```python
#importing required Libraries
import pandas as pd  #to create dataframe
import requests #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML
import numpy as np # to count the values
```

```python
#assigning the URL
url = 'https://www.imdb.com/search/title/?sort=user_rating,asc&groups=top_1000&count=100'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'}  #adding a User-Agent header to mimic a regular web browser
#request allow you to send HTTP request
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(response)
```

```python
#creating an empty list, so that we can append the values
movie_name = []
years = []
time = []
rating = []
metaScore = []
votes = []
```

```python
#storing the meaningfull required data in the variable movie_data
movie_data = soup.findAll('div', attrs={'class': 'sc-d80c3c78-4 kXzHjH dli-parent'})
```

```python
#to access the value after '.'
def get_after_dot(string):
    parts = string.split('.', 1)  # Split into two parts at the first occurrence of '.'
    if len(parts) > 1:
        return parts[1]  # Return the part after the '.'
    else:
        return ""
```

```python
for store in movie_data:
    #Movie_name
    name = store.h3.text
    movie_name.append(get_after_dot(name));

    #Year of Relase
    year_of_release = store.find_all('span', class_ = 'sc-b0691f29-8 ilsLEX dli-title-metadata-item')[0].text
    years.append( year_of_release);

    #Watchtime
    runtime = store.find_all('span', class_ = 'sc-b0691f29-8 ilsLEX dli-title-metadata-item')[1].text
    time.append(runtime)

    #Movie Rating
    rate = store.find('span', class_ = 'ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating').attrs['aria-label'].replace('IMDb rating: ', '')
    rating.append(rate)

    #Metascore
    meta = store.find('span', class_ = 'sc-b0901df4-0 bcQdDJ metacritic-score-box').text if store.find('span', class_ = 'sc-b0901df4-0 bcQdDJ metacritic-score-box') else '***'
    metaScore.append(meta)

    #Votes
    fetch_vote = store.find('div', class_ = 'sc-d80c3c78-0 jPoaKS').text.replace('Votes', '')
    votes.append(fetch_vote)
```

```python
#creating a dataframe using pandas library
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of relase': years, 'Watchtime': time, 'Movie Rating': rating, 'Metascore': metaScore, 'Votes': votes})
```

```python
#to start indexing from 1 instead of 0.
movie_DF.reset_index(drop=True, inplace=True)
movie_DF.index += 1
```
